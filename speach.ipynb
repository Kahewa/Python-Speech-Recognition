```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
frequency_sampling, audio_signal = wavfile.read("C:\\Users\\p.shuuya\\Downloads\\926a-f6b4-4eb7-8278-e0ca41171923.wav")
```

```python
print('\nSignal shape:', audio_signal.shape)
print('Signal Datatype:', audio_signal.dtype)
print('Signal duration:', round(audio_signal.shape[0] / 
float(frequency_sampling), 2), 'seconds')
```

```python
audio_signal = audio_signal / np.power(2, 15)
```

```python
audio_signal = audio_signal [:100]
time_axis = 1000 * np.arange(0, len(audio_signal), 1) / float(frequency_sampling)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming audio_signal is a stereo signal (388551 samples with 2 channels)
channel_1 = audio_signal[:, 0]  # First channel
channel_2 = audio_signal[:, 1]  # Second channel

# Generate time axis (assuming frequency_sampling is defined)
time_axis = 1000 * np.arange(0, len(channel_1), 1) / float(frequency_sampling)

# Plot each channel separately
plt.figure(figsize=(10, 6))
plt.plot(time_axis, channel_1, color='blue', label='Channel 1')
plt.plot(time_axis, channel_2, color='red', label='Channel 2')
plt.xlabel('Time (milliseconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
